```python
#these are all the imports you need
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
```

```python
x = np.linspace(-15,25,200)

N_0 = stats.norm.pdf(x, loc =0, scale =1)
N_10 = stats.norm.pdf(x, loc =10, scale =1)

t_0 = stats.t.pdf(x, 2, loc =0, scale =1)
t_10 = stats.t.pdf(x, 2, loc =10, scale =1)
```

```python
plt.plot(x, N_0, c = 'k', linestyle = 'solid', label = 'N(0,1)')
plt.plot(x, N_10, c = 'k', linestyle = 'dashed', label = 'N(0,10)')
plt.plot(x, t_0, c = 'b', linestyle = 'solid', label = 't(0,1)')
plt.plot(x, t_10, c = 'b', linestyle = 'dashed', label = 't(0,10)')
plt.legend(loc = 'upper right')
plt.xlabel('x')
plt.ylabel('not normalized distribution')

plt.show()
```

```python
def posterior(x, prior, likelihood):
    posterior = prior * likelihood

    plt.plot(x, prior/np.max(prior), c = 'k', label = 'prior')
    plt.plot(x, likelihood/np.max(likelihood), c = 'b', label = 'likelihood')
    plt.plot(x, posterior/np.max(posterior), c = 'r', label = 'posterior')
    plt.legend(loc = 'upper right')
    plt.xlabel('x')
    plt.ylabel('not normalized distribution')
    plt.show()
```

```python
posterior(x, N_0, N_10)
```

```python
posterior(x, N_0, t_0)
```

```python
posterior(x, N_0, t_10)
```

```python
posterior(x, t_0, N_10)
```

```python
posterior(x, t_0, t_10)
```
