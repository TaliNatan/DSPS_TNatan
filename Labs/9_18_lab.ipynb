```python
#these are all the imports you need
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
```

```python
x = np.linspace(-20,30,200)

N_0 = stats.norm.pdf(x, loc =0, scale =1)
N_10 = stats.norm.pdf(x, loc =10, scale =1)

t_0 = stats.t.pdf(x, 2, loc =0, scale =1)
t_10 = stats.t.pdf(x, 2, loc =10, scale =1)
```

```python
plt.plot(N_0)
plt.plot(N_10)
plt.plot(t_0)
plt.plot(t_10)

plt.show()
```
